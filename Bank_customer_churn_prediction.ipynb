**ELIMINATING THE UNWANTED COLUMNS**

```python
customer_churn = customer_churn.drop(['RowNumber','CustomerId', 'Surname'],axis = 1)
```

```python
customer_churn.head()
```

```python
customer_churn['Geography'].unique()
```

```python
customer_churn = pd.get_dummies(customer_churn,drop_first=True)
```

```python
customer_churn.head()
```

**UNBALANCED DATA**

```python
customer_churn['Exited'].value_counts()
```

```python
X = customer_churn.drop('Exited',axis = 1)
```

```python
y = customer_churn['Exited']
```

```python
y
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res = SMOTE().fit_resample(X,y)
```

**BALANCED DATA**

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_res,y_res,test_size=0.2,random_state=42)
```

```python
print(X_res.shape,X_train.shape,X_test.shape)
```

```python
print(y_res.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

```python
X_train
```

**MODEL TRAINING**

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
churn = LogisticRegression()
```

```python
churn.fit(X_train,y_train)
```

```python
X_pred = churn.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train,X_pred)
```

```python
y_pred = churn.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
